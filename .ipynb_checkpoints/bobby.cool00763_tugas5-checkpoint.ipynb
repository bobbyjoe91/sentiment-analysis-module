# Analisis Data Chinook Database
Chinook Database adalah basis data penjualan album dan *track* musik di toko Chinook. Basis data penjualan album terdiri atas 11 tabel yang merekam aktivitas penjualan dari tahun 2009-2013. Analisis basis data Chinook akan dilakukan secara visual melalui plot dan tabel basis data. Analisis ini berfokus pada 3 pertanyaan bisnis:

1. Bagaimana keadaan penjualan selama tahun 2013?
2. Apa genre favorit tiap negara selama tahun 2013?
3. Per bulannya, genre apa yang memiliki penjualan tertinggi sepanjang 2013?

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sqlite3
```

```python
%matplotlib inline
```

```python
# Menghubungkan ke Database
DB_PATH = "../../../../ChinookDB/chinook.sqlite"
try:
    connection = sqlite3.connect(DB_PATH)
    cursor = connection.cursor()
except:
    print(f'Database does not exist in {DB_PATH}')
```

### 1. Bagaimana keadaan penjualan selama 2013?

```python
select_invoice = '''
    SELECT strftime("%m/%Y", Inv.InvoiceDate) AS Period, SUM(Inv.Total) AS 'Total Sum' 
    FROM Invoice Inv
    WHERE strftime("%Y", Inv.InvoiceDate) = '2013'
    GROUP BY Period
    ORDER BY date(Inv.InvoiceDate) ASC;
'''

penjualan_per_bulan = pd.read_sql_query(select_invoice, connection)
summary = penjualan_per_bulan.describe().to_dict()['Total Sum']
summary
```

```python
periode = penjualan_per_bulan["Period"]
jumlah_penjualan = penjualan_per_bulan["Total Sum"]

sns.set(rc={'figure.figsize':(20,6)})
plot1 = sns.lineplot(x=periode, y=jumlah_penjualan, linewidth=3)
plot1.set_title("1. Penjualan tahun 2013", fontsize=15)
plt.show()
```

```python
import scipy.stats

data_penjualan_2013 = penjualan_per_bulan['Total Sum'].values
dist_penjualan_2013 = scipy.stats.norm.pdf(data_penjualan_2013, summary['mean'], summary['std'])

sns.kdeplot(data_penjualan_2013, shade=True).set_title("2. Distribusi Penjualan tahun 2013", fontsize=15)
plt.show()
```

Penjualan selama tahun 2013 cenderung stabil di sekitar 37 dengan kenaikan signifikan di bulan November 2013. Hal ini dapat dilihat pada plot 1 dan plot 2. Pada plot 2, terlihat distribusi penjualan berpusat di sekitar 37 dan terdistribusi normal. Nilai penjualan yang lebih rendah atau lebih tinggi dari 37 dapat dikatakan sebagai kejadian langka jika dilihat dari nilai <i>density</i>-nya. <br><br>Tren penjualan dapat dikatakan mengalami peningkatan, dimulai dari rendahnya penjualan di awal tahun dan diakhiri dengan tingginya penjualan di akhir tahun. Perlu informasi mengenai tren musik tiap bulan untuk menafsirkan informasi ini lebih lanjut.

### 2. Genre Favorit Tiap Negara (tahun 2013)

```python
favorite_genre_by_country = '''
    SELECT Inv.BillingCountry AS Country, Art.Name AS Name, Ge.Name AS Genre, COUNT(Art.Name) AS Artist_Count
    FROM
        Artist Art 
            INNER JOIN 
        Album Al ON Art.ArtistId = Al.ArtistId
            INNER JOIN
        Track Tr ON Al.AlbumId = Tr.AlbumId
            INNER JOIN
        Genre Ge ON Tr.GenreId = Ge.GenreId
            INNER JOIN
        InvoiceLine InvL ON Tr.TrackId = InvL.TrackId
            INNER JOIN
        Invoice Inv ON InvL.InvoiceId = Inv.InvoiceId
    WHERE strftime("%Y", Inv.InvoiceDate) = '2013'
    GROUP BY Country
    ORDER BY Artist_Count DESC;
'''
data_genre_favorit = pd.read_sql_query(favorite_genre_by_country, connection)
data_genre_favorit
```

Data genre terfavorit (2013) untuk tiap negara cenderung beragam, tetapi didominasi oleh artis Rock dan Latin. Data ini memungkinkan kita untuk melihat tren genre dan artis yang diminati di tahun 2013 sehingga kita bisa menjual album atau *track* dari genre tersebut lebih banyak di negara peminatnya kedepan.

### 3. Genre dengan Penjualan Tertinggi per Bulan (2013)

```python
max_sales_by_genre = '''
    SELECT Ge.Name AS Genre, MAX(Inv.Total) AS Max_Sales, strftime("%m/%Y", Inv.InvoiceDate) AS Date
    FROM
        Artist Art 
            INNER JOIN 
        Album Al ON Art.ArtistId = Al.ArtistId
            INNER JOIN
        Track Tr ON Al.AlbumId = Tr.AlbumId
            INNER JOIN
        Genre Ge ON Tr.GenreId = Ge.GenreId
            INNER JOIN
        InvoiceLine InvL ON Tr.TrackId = InvL.TrackId
            INNER JOIN
        Invoice Inv ON InvL.InvoiceId = Inv.InvoiceId
    WHERE strftime("%Y", Inv.InvoiceDate) = '2013'
    GROUP BY strftime("%m/%Y", Inv.InvoiceDate)
    ORDER BY date(Inv.InvoiceDate);
'''
penjualan_maks_genre = pd.read_sql_query(max_sales_by_genre, connection)
```

```python
plot3 = sns.barplot(x='Date', y='Max_Sales', 
            data=penjualan_maks_genre, 
            hue='Genre', dodge=False,
            palette=sns.color_palette("Paired"))

plot3.set_title("3. Genre dengan Penjualan Tertinggi tahun 2013", fontsize=15)
plt.show()
```

Analisis ini menunjukkan genre musik dengan penjualan terbesar setiap bulannya (2013). Musik Rock mendominasi penjualan di setiap bulan sepanjang tahun 2013. Pada bulan November 2013, musik Alternatif dan Punk menjadi genre yang paling diminati. Kedepannya, kita dapat meningkatkan penjualan musik genre Rock dan berfokus menarget penggemar genre Rock.

### Kesimpulan
Selama 2013, penjualan cenderung stabil namun terjadi penurunan dan kenaikan signifikan di awal dan akhir tahun. Penjualan bisa difokuskan ke pasar dan penggemar musik Rock di seluruh dunia karena pangsa pasarnya yang besar.
